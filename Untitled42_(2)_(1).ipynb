```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import MiniBatchKMeans
import time
from sklearn.cluster import DBSCAN
```

```python
mall_customers = pd.read_csv('Mall_Customers.csv')
print(mall_customers.head())
print(mall_customers.shape)
print(mall_customers.describe())
print(mall_customers.info())
```

```python
missing_values = mall_customers.isnull().sum()
missing_values[missing_values > 0]
```

```python
continuous_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
mall_customers[continuous_cols] = imputer.fit_transform(mall_customers[continuous_cols])
mall_customers
```

```python
missing_values = mall_customers.isnull().sum()
missing_values[missing_values > 0]
```

```python
gender_counts = mall_customers['Gender'].value_counts()
print(gender_counts)
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=['Female (0)', 'Male (1)'], autopct='%1.2f%%', colors=['skyblue', 'red'])
plt.title('Distribution of Transaction Classes')
```

```python
gender_counts = gender_counts.to_frame()
features = [col for col in gender_counts.columns if col != 'Gender']
gender_counts[features].hist(bins=50, figsize=(20,15), color='skyblue')
plt.suptitle("Feature Distributions BEFORE Scaling")
plt.tight_layout()

```

```python
Mall = mall_customers.drop(columns=['Gender'])
scaler = StandardScaler()
Mall_scaled = scaler.fit_transform(Mall)
Mall_scaled
```

```python
Mall_scaled_df = pd.DataFrame(Mall_scaled, columns=Mall.columns)
Mall_scaled_df.hist(bins=50, figsize=(20,15), color='pink')
plt.suptitle("Feature Distributions AFTER Scaling")
plt.tight_layout()
```

```python
pca = PCA()
Mall_pca = pca.fit_transform(mall_customers[continuous_cols])
Mall_pca.shape
```

```python
explained_var_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_var_ratio)
for i, (ev, cv) in enumerate(zip(explained_var_ratio, cumulative_variance), start=1):
    print(f"Component {i}: Explained Variance = {ev:.4f}, Cumulative = {cv:.4f}")
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% threshold')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% threshold')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components - Mall Customers')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1

print(f" 90% variance: {n_components_90}")
print(f" 95% variance: {n_components_95}")
```

```python
pca_2d = pca.transform(mall_customers[continuous_cols])[:, :2]
gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
plt.figure(figsize=(8, 6))
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=gender_numeric, cmap='coolwarm', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Mall Customers (Colored by Gender)')
plt.colorbar(label='Gender (0=Female, 1=Male)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
tsne = TSNE(n_components=2, perplexity=10, max_iter=300, random_state=42)
Mall_tsne_2d = tsne.fit_transform(mall_customers[continuous_cols])
plt.figure(figsize=(8, 6))
scatter = plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=gender_numeric, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('2D t-SNE Scatter Plot (Colored by Gender)')
cbar = plt.colorbar(scatter)
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Female', 'Male'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
plt.figure(figsize=(8, 6))
plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=gender_numeric, cmap='viridis', edgecolor='k', s=50)

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Before Clustering (Colored by Actual Gender)')
plt.grid(True)
cbar = plt.colorbar()
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Female', 'Male'])
plt.tight_layout()
plt.show()
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(mall_customers[continuous_cols])
plt.figure(figsize=(8, 6))
plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=clusters, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE After Clustering (Colored by K-Means Clusters)')
plt.grid(True)
plt.colorbar(label='Cluster Label')
plt.tight_layout()
plt.show()
```

```python


inertia = []
silhouette_scores = []
k_values = range(2, 20)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(mall_customers[continuous_cols])

    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(mall_customers[continuous_cols], cluster_labels)
    silhouette_scores.append(silhouette_avg)
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method For Optimal k (Mall Customers)')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Different k (Mall Customers)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
initializations = ['random', 'k-means++']
k_values = range(2, 20)

init_inertia = {}
init_silhouette = {}

for init in initializations:
    inertia = []
    silhouette_scores = []

    for k in k_values:
        kmeans = KMeans(n_clusters=k, init=init, random_state=42)
        cluster_labels = kmeans.fit_predict(mall_customers[continuous_cols])

        inertia.append(kmeans.inertia_)
        silhouette_avg = silhouette_score(mall_customers[continuous_cols], cluster_labels)
        silhouette_scores.append(silhouette_avg)

    init_inertia[init] = inertia
    init_silhouette[init] = silhouette_scores
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for init in initializations:
    plt.plot(k_values, init_inertia[init], label=f'{init}')
plt.title('Inertia Comparison: KMeans vs KMeans++')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
for init in initializations:
    plt.plot(k_values, init_silhouette[init], label=f'{init}')
plt.title('Silhouette Score Comparison: KMeans vs KMeans++')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python


k_values = range(2, 20)
kmeans_inertia = []
kmeans_silhouette = []
kmeans_time = []

minibatch_inertia = []
minibatch_silhouette = []
minibatch_time = []

for k in k_values:

    start_time = time.time()
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans_labels = kmeans.fit_predict(mall_customers[continuous_cols])
    end_time = time.time()
    kmeans_inertia.append(kmeans.inertia_)
    kmeans_silhouette.append(silhouette_score(mall_customers[continuous_cols], kmeans_labels))
    kmeans_time.append(end_time - start_time)

    start_time = time.time()
    minibatch = MiniBatchKMeans(n_clusters=k, init='k-means++', random_state=42, batch_size=20)
    minibatch_labels = minibatch.fit_predict(mall_customers[continuous_cols])
    end_time = time.time()
    minibatch_inertia.append(minibatch.inertia_)
    minibatch_silhouette.append(silhouette_score(mall_customers[continuous_cols], minibatch_labels))
    minibatch_time.append(end_time - start_time)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(k_values, kmeans_inertia, 'o-', label='KMeans')
plt.plot(k_values, minibatch_inertia, 's-', label='MiniBatchKMeans')
plt.title('Inertia Comparison')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.grid(True)
plt.legend()
plt.subplot(1, 3, 2)
plt.plot(k_values, kmeans_silhouette, 'o-', label='KMeans')
plt.plot(k_values, minibatch_silhouette, 's-', label='MiniBatchKMeans')
plt.title('Silhouette Score Comparison')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.legend()
plt.subplot(1, 3, 3)
plt.plot(k_values, kmeans_time, 'o-', label='KMeans')
plt.plot(k_values, minibatch_time, 's-', label='MiniBatchKMeans')
plt.title('Execution Time Comparison')
plt.xlabel('k')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(mall_customers[continuous_cols])

eps = 0.5
min_samples = 5

dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(X_scaled)

n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = list(dbscan_labels).count(-1)

print(f"Number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {n_noise}")
plt.figure(figsize=(8, 6))
plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1], c=dbscan_labels, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples})')
plt.grid(True)
plt.colorbar(label='Cluster Label')
plt.tight_layout()
plt.show()

```

```python

eps_values = [0.3, 0.5, 0.7, 1.0]
min_samples_values = [3, 5, 10]

best_config = {'score': -1, 'eps': None, 'min_samples': None}

for eps in eps_values:
    for min_s in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_s)
        labels = dbscan.fit_predict(X_scaled)

        if len(set(labels)) > 1 and len(set(labels)) < len(X_scaled):
            score = silhouette_score(X_scaled, labels)
            print(f"eps={eps}, min_samples={min_s}, silhouette score={score:.4f}")
            if score > best_config['score']:
                best_config = {'score': score, 'eps': eps, 'min_samples': min_s}

print(f"\nBest DBSCAN configuration:")
print(f"eps = {best_config['eps']}")
print(f"min_samples = {best_config['min_samples']}")
print(f"silhouette score = {best_config['score']:.4f}")
```

```python

best_k = 5

start = time.time()
kmeans = KMeans(n_clusters=best_k, init='k-means++', random_state=42)
kmeans_labels = kmeans.fit_predict(mall_customers[continuous_cols])
kmeans_inertia = kmeans.inertia_
kmeans_silhouette = silhouette_score(mall_customers[continuous_cols], kmeans_labels)
kmeans_time = time.time() - start
kmeans_clusters = best_k


start = time.time()
minibatch = MiniBatchKMeans(n_clusters=best_k, init='k-means++', random_state=42, batch_size=20)
minibatch_labels = minibatch.fit_predict(mall_customers[continuous_cols])
minibatch_inertia = minibatch.inertia_
minibatch_silhouette = silhouette_score(mall_customers[continuous_cols], minibatch_labels)
minibatch_time = time.time() - start
minibatch_clusters = best_k


best_eps = 0.5
best_min_samples = 5

start = time.time()
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
dbscan_labels = dbscan.fit_predict(X_scaled)
dbscan_time = time.time() - start
dbscan_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
dbscan_noise = list(dbscan_labels).count(-1)

if len(set(dbscan_labels)) > 1:
    dbscan_silhouette = silhouette_score(X_scaled, dbscan_labels)
else:
    dbscan_silhouette = None

comparison_data = {
    'Algorithm': ['KMeans', 'MiniBatchKMeans', 'DBSCAN'],
    'Inertia': [kmeans_inertia, minibatch_inertia, None],
    'Silhouette Score': [kmeans_silhouette, minibatch_silhouette, dbscan_silhouette],
    'Execution Time (s)': [kmeans_time, minibatch_time, dbscan_time],
    'Number of Clusters': [kmeans_clusters, minibatch_clusters, dbscan_clusters],
    'Noise Points (if any)': [None, None, dbscan_noise]
}

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df)
```
