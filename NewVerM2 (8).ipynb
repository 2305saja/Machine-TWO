```python
print("Before Undersampling :")
print(Card_Data['Class'].value_counts())

Card_Data_majority = Card_Data[Card_Data['Class'] == 0]
CD_minority = Card_Data[Card_Data['Class'] == 1]

CD_majority_sampled = CD_minority.sample(n=len(CD_minority), random_state=42)

CD_balanced_undersample = pd.concat([CD_majority_sampled, CD_minority])

df_balanced_undersample = CD_balanced_undersample.sample(frac=1, random_state=42).reset_index(drop=True)

print("After Undersampling:\n")
print(df_balanced_undersample['Class'].value_counts())

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
scatter = plt.scatter(
    df_balanced_undersample.iloc[:, 0],
    df_balanced_undersample.iloc[:, 1],
    c=df_balanced_undersample['Class'],
    cmap='spring',
    alpha=0.8
)
plt.title("Before PCA of Credit Card Transactions")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
cbar = plt.colorbar(scatter)
cbar.set_label("Class")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca_card=PCA(n_components=24)
Card_Data_pca=pca_card.fit_transform(df_balanced_undersample)
Card_Data_pca = pd.DataFrame(Card_Data_pca)
Card_Data_pca.head()
```

```python
plt.figure(figsize=(8, 8))
scatter = plt.scatter(
    Card_Data_pca.iloc[:, 0],
    Card_Data_pca.iloc[:, 1],
    c=df_balanced_undersample['Class'],
    alpha=0.8
)
plt.title("After PCA of Credit Card Transactions")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
cbar = plt.colorbar(scatter)
cbar.set_label("Class")
plt.show()
```

```python
X_meaned = Card_Data - np.mean(Card_Data, axis=0)
X_std = X_meaned / np.std(X_meaned, axis=0)

cov_matrix = np.cov(X_std, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]

explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

cumulative_explained_variance = np.cumsum(explained_variance_ratio)

for i, (var, cum_var) in enumerate(zip(explained_variance_ratio, cumulative_explained_variance), start=1):
    print(f"PC{i}: Explained Variance Ratio = {var:.4f}, Cumulative = {cum_var:.4f}")
```

```python
threshold = 0.90
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

print(f"Number of components to preserve {threshold*100}% variance: {num_components}")
```

```python
threshold = 0.95
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

print(f"Number of components to preserve {threshold*100}% variance: {num_components}")
```

##2. Clustering Analysis##

```python
from sklearn.cluster import KMeans

Inertia_Value = []
for i in range(2,20):
  kmean=KMeans(n_clusters=i,random_state=42)
  kmean.fit(Card_Data_pca)
  Inertia_Value.append(kmean.inertia_)
  print(f"Inertia for {i} clusters: {kmean.inertia_}")
```

```python
from sklearn.metrics import silhouette_score

labels = kmean.fit_predict(Card_Data_pca)

score = silhouette_score(Card_Data_pca, labels)
print("Silhouette Score:", score)
```

```python
silhouette_scores = []
k_values = range(2, 20)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(Card_Data_pca)
    score = silhouette_score(Card_Data_pca, labels)
    silhouette_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(k_values, Inertia_Value, marker='o', color='skyblue')
ax1.set_title('Elbow Method: Inertia vs. Number of Clusters (k)')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_xticks(k_values)
ax1.grid(True)

ax2.plot(k_values, silhouette_scores, marker='o', color='orange')
ax2.set_title('Silhouette Score vs. Number of Clusters (k)')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.grid(True)

plt.tight_layout()
plt.show()

```
