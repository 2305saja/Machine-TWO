```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
mall_customers = pd.read_csv('Mall_Customers.csv')
print(mall_customers.head())
print(mall_customers.shape)
print(mall_customers.describe())
print(mall_customers.info())
```

```python
missing_values = mall_customers.isnull().sum()
missing_values[missing_values > 0]
```

```python
continuous_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
mall_customers[continuous_cols] = imputer.fit_transform(mall_customers[continuous_cols])
mall_customers
```

```python
missing_values = mall_customers.isnull().sum()
missing_values[missing_values > 0]
```

```python
gender_counts = mall_customers['Gender'].value_counts()
print(gender_counts)
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=['Normal (0)', 'Fraud (1)'], autopct='%1.2f%%', colors=['skyblue', 'red'])
plt.title('Distribution of Transaction Classes')
```

```python
gender_counts = gender_counts.to_frame()
features = [col for col in gender_counts.columns if col != 'Gender']
gender_counts[features].hist(bins=50, figsize=(20,15), color='skyblue')
plt.suptitle("Feature Distributions BEFORE Scaling")
plt.tight_layout()

```

```python
Mall = mall_customers.drop(columns=['Gender'])
scaler = StandardScaler()
Mall_scaled = scaler.fit_transform(Mall)
Mall_scaled
```

```python
Mall_scaled_df = pd.DataFrame(Mall_scaled, columns=Mall.columns)
Mall_scaled_df.hist(bins=50, figsize=(20,15), color='pink')
plt.suptitle("Feature Distributions AFTER Scaling")
plt.tight_layout()
```

```python
pca = PCA()
Mall_pca = pca.fit_transform(mall_customers[continuous_cols])
Mall_pca.shape
```

```python
explained_var_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_var_ratio)
for i, (ev, cv) in enumerate(zip(explained_var_ratio, cumulative_variance), start=1):
    print(f"Component {i}: Explained Variance = {ev:.4f}, Cumulative = {cv:.4f}")
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% threshold')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% threshold')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components - Mall Customers')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1

print(f" 90% variance: {n_components_90}")
print(f" 95% variance: {n_components_95}")
```

```python
pca_2d = pca.transform(mall_customers[continuous_cols])[:, :2]
gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
plt.figure(figsize=(8, 6))
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=gender_numeric, cmap='coolwarm', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Mall Customers (Colored by Gender)')
plt.colorbar(label='Gender (0=Female, 1=Male)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
tsne = TSNE(n_components=2, perplexity=10, max_iter=300, random_state=42)
Mall_tsne_2d = tsne.fit_transform(mall_customers[continuous_cols])
plt.figure(figsize=(8, 6))
scatter = plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=gender_numeric, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('2D t-SNE Scatter Plot (Colored by Gender)')
cbar = plt.colorbar(scatter)
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Female', 'Male'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

gender_numeric = mall_customers['Gender'].map({'Female': 0, 'Male': 1})
plt.figure(figsize=(8, 6))
plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=gender_numeric, cmap='viridis', edgecolor='k', s=50)

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Before Clustering (Colored by Actual Gender)')
plt.grid(True)
cbar = plt.colorbar()
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Female', 'Male'])
plt.tight_layout()
plt.show()
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(mall_customers[continuous_cols])
plt.figure(figsize=(8, 6))
plt.scatter(Mall_tsne_2d[:, 0], Mall_tsne_2d[:, 1],c=clusters, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE After Clustering (Colored by K-Means Clusters)')
plt.grid(True)
plt.colorbar(label='Cluster Label')
plt.tight_layout()
plt.show()
```

```python


inertia = []
silhouette_scores = []
k_values = range(2, 20)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(mall_customers[continuous_cols])

    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(mall_customers[continuous_cols], cluster_labels)
    silhouette_scores.append(silhouette_avg)
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method For Optimal k (Mall Customers)')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Different k (Mall Customers)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
